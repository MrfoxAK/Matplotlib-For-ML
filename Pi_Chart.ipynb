# Pie Chart

```python
import matplotlib.pyplot as plt
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie([1])
plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students)
plt.show()
```

```python
explode = [0.02,0,0.1,0,0]
colors = ["c", 'b','r','y','g']
plt.pie(class1_students, labels = classes,explode = explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4)
plt.show()
```

```python
explode = [0.02,0,0.1,0,0.2]
colors = ["c", 'b','r','y','g']
textprops = {"fontsize":15}
plt.pie(class1_students, labels = classes,explode = explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=90,textprops=textprops)
plt.show()
```
