# plt.imshow & Colorbar

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
from matplotlib import style
import pandas as pd
```

```python
img1 = img.imread('subplot_figure.png')
```

```python
plt.imshow(img1)
plt.show()
```

```python
plt.imshow(img1)
plt.colorbar()
plt.show()
```

```python
plt.imshow(img1, cmap='hot')
plt.colorbar()
plt.show()
```
