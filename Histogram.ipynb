# Histogram 

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students = np.random.randint(18,45,(100))
ds_students = np.random.randint(15,40,(100))
print(ml_students)
print(ds_students)
```

```python
plt.hist(ml_students)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
 
plt.hist(ml_students, bins, rwidth=0.8, histtype = "bar",
         orientation='horizontal', color = "m", label = "ML Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
 
plt.hist(ml_students, bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = "m", label = "ML Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
 
plt.hist(ml_students, bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = "m", label = "ML Student")
plt.hist(ds_students, bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = "y", label = "DS Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize = (16,9)) # size of histogram in 16:9 format
 
 
plt.hist([ml_students,ds_students], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m","y"], label = ["ML Student","DS Student"])
# plt.hist(ds_students, bins, rwidth=0.8, histtype = "bar",
#          orientation='vertical', color = "y", label = "DS Student")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
