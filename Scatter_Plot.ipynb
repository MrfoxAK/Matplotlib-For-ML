```python
df = pd.read_csv("googleplaystore.csv",nrows=100)
```

```python
df
```

```python
x = df["Rating"]
y = df["Reviews"]
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y , c='black', marker='*',s=120)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# alpha stands for transfarency
plt.figure(figsize = (16,9))
plt.scatter(x,y , c='black', marker='*',s=120, alpha=0.25)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y , c='black', marker='*',s=120, alpha=0.25)
plt.scatter(x,df["Installs"] , c='y', marker='o',s=120)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Install")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
 
plt.scatter(x,df["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, linewidths=10,
           edgecolors="c" )
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```
